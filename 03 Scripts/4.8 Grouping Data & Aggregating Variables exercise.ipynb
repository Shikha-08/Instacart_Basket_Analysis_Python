# Step 4 Creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
# Aggregating Data with transform()

ords_prods_merge['max_order'] = ords_prods_merge.groupby('user_id')['order_number'].transform('max')
```

```python
ords_prods_merge.head(20)
```

```python
# Deriving Columns with loc()

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna= False)
```

```python
# updated output

ords_prods_merge.head(20)
```

# Step 5  Analyze whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices' : ['mean', 'min', 'max']})
```

New customers tend to spend more on average per purchase, while loyal customers spend the least. All groups share the same minimum (1.0) and maximum (99,999.0) price points.

# Step 6 Create a spending flag for each user based on the average price across all their orders using given criteria.

```python
# Aggregating Data with transform()

ords_prods_merge['avg_product_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
```

```python
ords_prods_merge.head()
```

```python
# Deriving "Spending_flag "Columns with loc()

ords_prods_merge.loc[ords_prods_merge['avg_product_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['avg_product_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna= False)
```

```python
ords_prods_merge[['user_id', 'spending_flag', 'avg_product_price']].head(10)
```

# Step 7 Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column.

```python
# Aggregating Data with transform()

ords_prods_merge['user_ordering_behavior'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
ords_prods_merge.head()
```

```python
# Deriving "frequency_flag "Columns with loc()

ords_prods_merge.loc[ords_prods_merge['user_ordering_behavior'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['user_ordering_behavior'] <= 20) & (ords_prods_merge['user_ordering_behavior'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['user_ordering_behavior'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```

```python
# Checking Relevant Columns of Updated Dataframe

ords_prods_merge[['user_id', 'days_since_prior_order', 'user_ordering_behavior', 'frequency_flag']].head(20)
```
