# 3. Provide the Instacart senior stakeholders with descriptive findings about sales. Create a histogram of the “order_hour_of_day” column.

```python
hist = df_instacart['order_hour_of_day'].plot.hist(bins = 24, alpha=0.7, title='Order Distribution by Hour of Day').set(xlabel='Hour of Day', ylabel='Frequency')
```

This chart highlights customer ordering behavior, showing that most orders are placed during daytime hours, while nighttime and early morning hours see significantly fewer orders.

# 7. Explore whether there’s a connection between age and spending power (income).

```python
# create a DataFrame with only the necessary columns

df_instacart_4 = small[['age','income']]
```

```python
df_instacart_4.head()
```

```python
# create a scatterplot

scatterplot = sns.scatterplot(data = df_instacart_4, x = 'age', y = 'income')
```

The scatterplot shows most incomes fall below $300,000 across all ages, with a dense cluster under $200,000. Higher incomes above $400,000 are rare and mostly outliers, with no clear relationship between age and income.
