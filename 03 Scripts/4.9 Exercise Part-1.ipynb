# 5. Complete the fundamental data quality and consistency checks youâ€™ve learned throughout this Achievement

```python
# Checking for mixed-type columns

for col in df_customer.columns.tolist():
  weird = (df_customer[[col]].map(type) != df_customer[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_customer[weird]) > 0:
    print (col)
```

Based on the above output, the "first name" column contains mixed data types, which need to be addressed.
