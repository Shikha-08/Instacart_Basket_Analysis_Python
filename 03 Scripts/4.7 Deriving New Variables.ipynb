```python
# define a function for price labels

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range product'
  else: return 'Not enough data'
```

```python
# apply the function

df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna=False)
```

```python
# Step 2  Created two busiest days of the week and two slowest days

result_updated = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value == 1:
    result_updated.append("Busiest days")
  elif value == 3 or value == 4:
    result_updated.append("Slowest days")
  else:
    result_updated.append("Regularly busy")
```

```python
result_updated
```

```python
# step 3 Created new column in ords_prods_merge to view results

ords_prods_merge['busiest days'] = result_updated
```

```python
ords_prods_merge['busiest days'].value_counts(dropna= False)
```

```python
# Check the values of this new column

ords_prods_merge.head()
```

The total count in ords_prods_merge matches the sum of all groupings, with each label's total aligning with its qualifying days. For example, "Busiest days" (11,864,412) matches orders on days 0 and 1 (6,204,182 + 5,660,230).

```python
# step 4 Create a new column containing these labels called “busiest_period_of_day.”

ords_prods_merge['order_hour_of_day'].value_counts(dropna=False)
```

The order_hour_of_day counts are split into: "Most orders" (10, 11, 14, 15, 13, 12, 16, 9), "Fewest orders" (23, 6, 0, 1, 5, 2, 4, 3), and "Average orders" (all remaining hours).

```python
result_for_hours_of_day = []

for value in ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result_for_hours_of_day.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    result_for_hours_of_day.append("Fewest orders")
  else:
    result_for_hours_of_day.append("Average orders")
```

```python
result_for_hours_of_day
```

```python
#  Created new column in ords_prods_merge to view results

ords_prods_merge['busiest_period_of_day'] = result_for_hours_of_day
```

```python
# Step 5 printing the frequency 

ords_prods_merge['busiest_period_of_day'].value_counts(dropna= False)
```
